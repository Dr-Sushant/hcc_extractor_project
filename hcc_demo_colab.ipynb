```python
%%writefile app.py
import streamlit as st

st.set_page_config(page_title="HCC Extractor", layout="centered")
st.title("🏥 HCC Code Extractor")
st.markdown("Extract Hierarchical Condition Category codes from clinical text")

# Simple mapping for demo
hcc_map = {
    "diabetes": "E11.9 → HCC18",
    "ckd": "N18.3 → HCC134",
    "hiv": "B20 → HCC01",
    "leukemia": "C92.00 → HCC08",
    "obesity": "E66.9 → HCC22",
    "heart failure": "I50.9 → HCC85",
}

clinical_text = st.text_area("Paste clinical note:", height=150,
                              placeholder="Example: Patient has type 2 diabetes and CKD stage 3")

if clinical_text:
    st.subheader("📋 Extracted Codes")
    text_lower = clinical_text.lower()
    found = False
    for condition, code in hcc_map.items():
        if condition in text_lower:
            st.success(f"✅ {condition.upper()}: {code}")
            found = True
    if not found:
        st.info("No HCC codes detected. Try: 'Patient has type 2 diabetes'")
```
